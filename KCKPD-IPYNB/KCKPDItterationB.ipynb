```python
import pandas as pd
import numpy as np

df = pd.read_excel("C://Users//hlcma//OneDrive//Desktop//KCKPD//KCKPD-YTD-STATS-1-15-24//KCKPD-YTD-STATS-1-15-24.xlsx", header=2)
## Header is of two rows with the Pivot table
## Reference Link: https://stackoverflow.com/questions/37400974/error-unicode-error-unicodeescape-codec-cant-decode-bytes-in-position-2-3
## Reference Link: https://saturncloud.io/blog/how-to-use-python-and-pandas-to-convert-xlsx-to-csv-and-remove-the-index-column/

df = df.astype(str) #convert all columns to strings as I have to combine numbers in the same cell
df = df.replace('nan','') #get rid of the nan created back to a blank string

# Combine 'Column1' and 'Column2' into a new column 'Combined'
df['2018'] = df[2018].astype(str) + df['Unnamed: 4']
df['2019'] = df[2019].astype(str) + df['Unnamed: 6']
df['2020'] = df[2020].astype(str) + df['Unnamed: 8']
df['2021'] = df[2021].astype(str) + df['Unnamed: 10']
df['2022'] = df[2022].astype(str) + df['Unnamed: 12']
df['2023'] = df[2023].astype(str) + df['Unnamed: 14']
df['2024'] = df[2024].astype(str) + df['Unnamed: 16']
df['(+/-)'] = df['(+/-)'].astype(str) + df['Unnamed: 18']
df['%'] = df['%'].astype(str) + df['Unnamed: 20']

df = df.drop(columns=[2018, 'Unnamed: 4', 2019, 'Unnamed: 6', 2020, 'Unnamed: 8', 2021, 'Unnamed: 10', 2022, 'Unnamed: 12', 2023, 'Unnamed: 14', 2024, 'Unnamed: 16', '(+/-)', 'Unnamed: 18', '%', 'Unnamed: 20'])

df_transposed = df.transpose()  # or df.transpose()
df_transposed
## Reference https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.transpose.html
```
